```python
import gradio as gr
from huggingface_hub import InferenceClient
from transformers import AutoTokenizer, AutoModelForCausalLM


model_name = "aaditya/Llama3-OpenBioLLM-70B"

#  path to save the model  on your Google Drive
save_path = "/content/drive/MyDrive/med_models/llama3_openbiollm_70b"

# Download and save the model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("aaditya/Llama3-OpenBioLLM-70B", cache_dir=save_path)
model = AutoModelForCausalLM.from_pretrained("aaditya/Llama3-OpenBioLLM-70B", cache_dir=save_path, device_map="auto", load_in_8bit=True)

# Load the model and tokenizer from the saved path
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", load_in_8bit=True)

def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=200)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

interface = gr.Interface(fn=generate_text, inputs="text", outputs="text")
interface.launch()
```

```python
import gradio as gr
from huggingface_hub import InferenceClient
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "aaditya/OpenBioLLM-Llama3-8B"

# Load the model and tokenizer (optional: reduce batch size)
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", batch_size=1)  # Reduce batch size to 1

def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=200)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

interface = gr.Interface(fn=generate_text, inputs="text", outputs="text")
interface.launch()
```
